```python
df['Avg. Area Number of Bedrooms'] = df['Avg. Area Number of Bedrooms'].apply( lambda x:round(x))
df.head(3)
```

### Modelling

```python
X = df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
xtrain, xtest, ytrain,  ytest = train_test_split(X,y,test_size=0.20,random_state = 101)
```

```python
model = LinearRegression().fit(xtrain,ytrain)
```

```python
round(model.score(xtest,ytest)*100,2)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_,columns=['Coef'],index= X.columns)
```

### Evaluation

```python
ypred = model.predict(xtest)
ypred
```

```python
from sklearn.metrics import mean_absolute_error as mae
```

```python
mae(ytest,ypred)
```

```python
100*mae(ytest,ypred)/ypred[0]
```
